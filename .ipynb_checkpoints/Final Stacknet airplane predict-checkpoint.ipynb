```python
def convert(predict):
    out = []
    output = []
    for i in range(len(predict)):
        temp = max(predict[i])
        for j in range(4):
            if temp == predict[i][j]:
                out.append(j)
    for i in out:
        if i==0:
            output.append("Highly_Fatal_And_Damaging")
        if i==1:
            output.append("Minor_Damage_And_Injuries")
        if i==2:
            output.append("Significant_Damage_And_Fatalities")
        if i==3:
            output.append("Significant_Damage_And_Serious_Injuries")
    return output
```

```python
def convert2(x):
    sample = []
    for i in x:
        if i=="Minor_Damage_And_Injuries":
            sample.append(1)
        if i=="Significant_Damage_And_Serious_Injuries":
            sample.append(3)
        if i=="Significant_Damage_And_Fatalities":
            sample.append(2)
        if i=="Highly_Fatal_And_Damaging":
            sample.append(0)
    return sample
```

```python
def convert3(x):
    sample = []
    for i in x:
        if i==1:
            sample.append("Minor_Damage_And_Injuries")
        if i==3:
            sample.append("Significant_Damage_And_Serious_Injuries")
        if i==2:
            sample.append("Significant_Damage_And_Fatalities")
        if i==0:
            sample.append("Highly_Fatal_And_Damaging")
    return sample
```

```python
def create_model(model):
    model.add(Dense(24,activation=tf.nn.relu ,kernel_initializer='lecun_uniform',kernel_regularizer = reg))
    model.add(BatchNormalization())
    model.add(Dense(4,activation=tf.nn.softmax))
```

```python
def objective_nn(params):
    cv = StratifiedKFold(4, random_state=42,shuffle = True)
    li = []
    for train, test in cv.split(normal,convert(y_nn.values)):
        X_train, X_test = normal[train], normal[test]
        y_train, y_test = y_nn.values[train], y_nn.values[test]
        reg = l2(float(params["regularizer"]))
        m = Sequential()
        create_model(m)
    
        adam = keras.optimizers.Adam(learning_rate=float(params["learning_rate"]), beta_1=0.9, beta_2=0.999, amsgrad=False)
        
        m.compile(optimizer=adam, loss='categorical_crossentropy',metrics=['categorical_crossentropy'])
        reduce_lr = ReduceLROnPlateau(monitor='categorical_crossentropy',mode = "min", factor=float(params["factor"]),patience=int(params["patience"]), min_lr=0.000001)
        history = m.fit(X_train,y_train, epochs=int(params["epochs"]),batch_size = int(params["batch_size"]),callbacks = [reduce_lr],verbose =0)
        nn_pred = m.predict_classes(X_test)
        li.append(f1_score(convert2(convert(y_test)),nn_pred,average = "weighted"))
        print(f1_score(convert2(convert(y_test)),nn_pred,average = "weighted"))
    output = (sum(li)/4)
    return output*(-1)
```

```python
# mean encoding
# this first dummifies the target then creates mean encodings.
def mean_encoding(columns,df,target,df_test): # here columns is the name of the columns to encode and df is the dataframe target is the column name for the target variable
    mean_en = pd.DataFrame()
    mean_en_test = pd.DataFrame()
    temp = pd.DataFrame(df[target])
    n = temp[target].nunique()
    un = temp[target].unique()
    count = 0
    new_target = pd.DataFrame()
    for i in range(n):
        tempx = []
        for j in range(df.shape[0]):
            if un[i] == temp.iloc[j,0]:
                tempx.append(1)
            else:
                tempx.append(0)
        new_target[un[i]] = tempx
    df = df.drop(target,axis = 1)
    temp2 = pd.concat([df,new_target],axis = 1)
    for i in columns:
        for j in range(n):
            temp = temp2.groupby(i)[un[j]].mean()
            mean_en[f"{i}+{un[j]}_mean_encoding"] = temp2[i].map(temp)
            mean_en_test[f"{i}+{un[j]}_mean_encoding"] = df_test[i].map(temp)
            if mean_en[f"{i}+{un[j]}_mean_encoding"].isnull().sum()>0:
                glob_mean = temp2[un[j]].mean()
                mean_en[f"{i}+{un[j]}_mean_encoding"] = mean_en[f"{i}+{un[j]}_mean_encoding"].fillna(glob_mean)
                mean_en_test[f"{i}+{un[j]}_mean_encoding"] = mean_en_test[f"{i}+{un[j]}_mean_encoding"].fillna(glob_mean)
    return mean_en,mean_en_test
```

```python
def train_rf(params,df,y):
    
    rf = RandomForestClassifier(n_jobs = -1,n_estimators = params["n_estimators"],max_depth = params["max_depth"])
    rf.fit(df,y)
    
    return rf
    
```
